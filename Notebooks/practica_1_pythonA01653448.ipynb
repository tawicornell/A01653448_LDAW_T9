### Función suma

$$
    suma(a) = \sum_{i=1}^{n}a_i
$$

````python
def suma(x):
    suma = 0
    #<rellena>
    for i in x:
      suma = suma + i
    return suma
````

### Función promedio.

$$
	\bar{x} = \dfrac{\sum_{i=1}^{n}x_i}{n}
$$

````python
def promedio(x):
    promedio = 0
    suma=0
    #<rellena>
    for i in x:
      suma = suma + i
    promedio = suma / len(x)
    
    return promedio
````

````python
assert promedio([1, 2, 3, 4, 5]) == 3, "Error en promedio"
````

### Función varianza.
$$
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n}(x_i - \bar{x})^2
$$

````python
def var(x):
    var = 0
    #<rellena>

    promedio = 0
    suma=0
    #<rellena>
    for i in x:
      suma = suma + i
    promedio = suma / len(x)

    suma2 = 0
    for i in x:
      suma2 = suma2 + (i-promedio)**2

    var = suma2 / len(x)
    
    return var
````

### Ones 2D

Completa la función ones_2d que recibe como parámetros un entero $n$ y un entero $m$ y retorna una lista con $n$ sub-listas cada una de tamaño $m$ con todos sus elementos con el valor 1. En otras palabras, una matriz $n \times m$ con sus elementos iniciados con valor $1$.

Ejemplo:



```
>>> ones_2d(2, 3)
[[1, 1, 1], [1, 1, 1]]
```



````python
def ones_2d(m, n):
    ones_list = []
    i = 0
    j = 0
    while i < m:
      ones_list.append([])

      while j < n:
        ones_list[i].append(1)
        j=j+1

      i=i+1
      j=0

    print(ones_list)
    return ones_list
````

### Producto punto (inner)

Completa la función ones_2d que recibe como parametros dos listas a, b de la misma longitúd y retorna el producto punto de ambos.

El producto punto está definido como:

$$a = [a_1 a_2 ... a_n] \: y \: b = [b_1 b_2 ... b_n]$$

$$
a \cdot b = a_1*b_1 + a_2*b_2 + \dots + a_n*b_n
$$

````python
def inner(a, b):
    #tu codigo
    componente = 0
    suma = 0
    resultado = 0
    n = 0

    while n < len(a):
      componente = a[n]*b[n]
      suma = suma + componente
      componente = 0
      n += 1  

    resultado = suma
    return resultado
````

````python
assert inner([1, 2, 3], [1, 2 ,3]) == 14, "Error en inner"
print("Correcto")
````

### Matriz identidad
$$
	I_{ij} = \begin{cases}
		1 & i=j\\
		0 & i \neq j\\
	\end{cases}
$$

````python
def eye(n):
    '''
    Esta funcion debe regresar la matriz identidad de tamano nxn
    '''
    ones_list = []
    i = 0
    j = 0
    escalonado = 0
    flag = False
    while i < n:
      ones_list.append([])

      while j < n:
        if escalonado == j:
          ones_list[i].append(1)
        else:
          ones_list[i].append(0)
        
        j=j+1
      escalonado += 1
      i=i+1
      j=0

    print(ones_list)

    return ones_list
````

### Matriz transpuesta

$$C = A^T \Leftrightarrow c_{ij} = a{ji}$$

Por ejemplo, si:
	$$
	A = \begin{bmatrix}
	a_{11} & a_{12}  \\
	a_{21} & a_{22}  \\
	a_{31} & a_{32}  \\
	\end{bmatrix}
	$$
	entonces
	$$
	A^T = \begin{bmatrix}
	a_{11} & a_{21} & a_{31} \\
	a_{12} & a_{22} & a_{32} \\
	\end{bmatrix}
	$$

````python
def traspose(A):
    '''
    Esta funcion debe regresar la transpuesta de A
    '''
 
    A_t = []
    for m in range(len(A[0])):
      l2 = []
      for n in range(len(A)):
        l2.append(A[n][m])
        
      A_t.append(l2)
 
    return A_t
````

````python
assert traspose([[1, 2], [3, 4], [5, 6]]) == [[1, 3, 5], [2, 4, 6]], "Error en traspose"
````

### Multiplicación de matrices

Si $A \in \mathbb{R}^{m \times n}$ y $B \in \mathbb{R}^{n \times p}$ entonces:
$$
C = A B \in \mathbb{R}^{m \times p}\Leftrightarrow c_{ij} = \sum_{k=1}^{n}a_{ik} b_{kj}
$$

````python
def mat_mul(A, B):

    a = len(B[0])
    b = len(A)
    C =[]
    if validate_mat_mul(A,B) == True:
      for i in range(b):
        l = []

        for i in range(a):
          l.append(0)

        C.append(l)

      for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                C[i][j] += A[i][k] * B[k][j]

    else:
      print("No se puede multiplicar")
    
    return C
    
````
